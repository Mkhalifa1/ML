```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x=np.array([[2,4],[3,3],[3,4],[5,4],[5,6],
            [5,8],[6,4],[6,5],[6,7],[7,3],[7,4],
            [8,2],[9,4],[10,6],[10,7],[10,8],[11,5],
            [11,8],[12,7],[13,6],[13,7],
            [14,6],[15,4],[15,5]])
```

```python
KMeans_cluster_2 = KMeans(init="random",n_clusters=2,n_init=10,max_iter=300,random_state=0)

print('K=2',KMeans_cluster_2.fit_predict(x) )

plt.scatter(x[:, 0], x[:, 1], c=KMeans_cluster_2.labels_, s=50, cmap='rainbow')

centers_2 = KMeans_cluster_2.cluster_centers_
plt.scatter(centers_2[:, 0], centers_2[:, 1], c='black', s=100, alpha=0.9);
plt.title('Clusters of K=2')
```

```python
KMeans_cluster_3 = KMeans(init="random",n_clusters=3,n_init=10,max_iter=300,random_state=0)

print('K=3',KMeans_cluster_3.fit_predict(x) )

plt.scatter(x[:, 0], x[:, 1], c=KMeans_cluster_3.labels_, s=50, cmap='rainbow')

centers_3 = KMeans_cluster_3.cluster_centers_
plt.scatter(centers_3[:, 0], centers_3[:, 1], c='black', s=100, alpha=0.9);
plt.title('Clusters of K=3')
```

```python
KMeans_cluster_4 = KMeans(init="random",n_clusters=4,n_init=10,max_iter=300,random_state=0)

print('K=4',KMeans_cluster_4.fit_predict(x) )

plt.scatter(x[:, 0], x[:, 1], c=KMeans_cluster_4.labels_, s=50, cmap='rainbow')

centers_4 = KMeans_cluster_4.cluster_centers_
plt.scatter(centers_4[:, 0], centers_4[:, 1], c='black', s=100, alpha=0.9);
plt.title('Clusters of K=4')
```
